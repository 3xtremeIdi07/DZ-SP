```python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x0):
    f = lambda x: 2 * np.sin(x) - np.cos(2*x)
    df = lambda x: 2 * np.cos(x) + 2 * np.sin(2*x)

    x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y_vals = f(x_vals)
    dy_vals = df(x_vals)

    y0 = f(x0)
    dy0 = df(x0)

    tangent_y = dy0 * (x_vals - x0) + y0
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, 'b-', lw=2, label='f(x) = 2sin(x) - cos(2x)')
    plt.plot(x_vals, dy_vals, 'g--', lw=2, label="f'(x) = 2cos(x) + 2sin(2x)")
    plt.plot(x_vals, tangent_y, 'r-', lw=1.5, label=f'Tangent at x = {x0:.2f}')
    plt.plot(x0, y0, 'ro', markersize=8)

    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.grid(True, alpha=0.3)
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-6, 8)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Function, its Derivative and Tangent Line at x = {x0:.2f}')
    plt.legend()
    plt.show()
```
