```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def inverse_exp_sample(lam, n_samples):
    U = np.random.rand(n_samples)
    X = -np.log(1 - U) / lam
    return X

lam = 2.0
n = 1000
sample = inverse_exp_sample(lam, n)

plt.hist(sample, bins=30, density=True, alpha=0.7, label='Sample')

x = np.linspace(0, max(sample), 200)
pdf = stats.expon.pdf(x, scale=1/lam)  # scale = 1/λ
plt.plot(x, pdf, 'r-', label='Theoretical PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('λ={lam}')
plt.legend()
plt.show()
```
